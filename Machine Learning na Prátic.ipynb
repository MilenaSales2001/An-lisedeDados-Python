<h1>Previsão de Compras com Árvore de Decisão: Um Guia para Iniciantes</h1>

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

# 1. Criando o DataFrame com dados simulados de clientes
dados = pd.DataFrame({
    'Idade': [25, 40, 35, 20],
    'Visitou_Pagina_Promocoes': [1, 0, 1, 0],  # 1 = Sim, 0 = Não
    'Comprou': ['Sim', 'Não', 'Sim', 'Não']
})

# 2. Separando variáveis de entrada (X) e saída (y)
X = dados[['Idade', 'Visitou_Pagina_Promocoes']]
y = dados['Comprou']

# 3. Treinando o modelo
modelo = DecisionTreeClassifier()
modelo.fit(X, y)

# 4. Visualizando a árvore
plt.figure(figsize=(8, 5))
plot_tree(modelo, feature_names=X.columns, class_names=modelo.classes_, filled=True)
plt.title("Árvore de Decisão - Compras Online")
plt.show()

# 5. Fazendo uma previsão com novo cliente
novo_cliente = pd.DataFrame({'Idade': [30], 'Visitou_Pagina_Promocoes': [1]})
predicao = modelo.predict(novo_cliente)

print(f" Previsão para novo cliente: {predicao[0]}")


```
