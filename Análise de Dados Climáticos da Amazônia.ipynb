<h1>Análise de Dados Climáticos da Amazônia<h1> 

```python
import pandas as pd

# Criando um DataFrame simulado com dados climáticos e de desmatamento
dados_amazonia = pd.DataFrame({
    'ano': [2015, 2016, 2017, 2018, 2019],
    'estado': ['AM', 'AM', 'AM', 'AM', 'AM'],
    'temperatura_media': [26.4, 26.7, 27.1, 27.3, 27.5],  # em °C
    'precipitacao_total': [2300, 2150, 2400, 2000, 1950],  # em mm
    'area_desmatada_km2': [5500, 6200, 6800, 7500, 9800]   # em km²
})

print(dados_amazonia)
```

```python
import matplotlib.pyplot as plt

plt.plot(dados_amazonia['ano'], dados_amazonia['temperatura_media'], marker='o', color='orange')
plt.title('Evolução da Temperatura Média - AM')
plt.xlabel('Ano')
plt.ylabel('Temperatura Média (°C)')
plt.grid(True)
plt.show()
```

```python
plt.plot(dados_amazonia['ano'], dados_amazonia['precipitacao_total'], marker='o', color='blue')
plt.title('Precipitação Total Anual - AM')
plt.xlabel('Ano')
plt.ylabel('Precipitação (mm)')
plt.grid(True)
plt.show()
```

```python
plt.plot(dados_amazonia['ano'], dados_amazonia['area_desmatada_km2'], marker='o', color='green')
plt.title('Área Desmatada - AM')
plt.xlabel('Ano')
plt.ylabel('Área Desmatada (km²)')
plt.grid(True)
plt.show()
```
