<h1>Curva do Conhecimento: O que os Números Dizem sobre Nossas Notas?</h1>

```python
import pandas as pd

# Criando um DataFrame
dados = {
    'Aluno': ['Ana', 'Bruno', 'Carlos', 'Diana', 'Eduardo'],
    'Nota_Prova_1': [8.0, 7.5, 9.0, 6.5, 8.5],
    'Nota_Prova_2': [7.0, 8.0, 9.5, 6.0, 9.0],
    'Faltas': [2, 0, 1, 3, 0]
}

df = pd.DataFrame(dados)

# Visualizando o DataFrame
print(df)
```

```python
# Média das provas
df['Média'] = (df['Nota_Prova_1'] + df['Nota_Prova_2']) / 2

# Estatísticas descritivas
estatisticas = df[['Nota_Prova_1', 'Nota_Prova_2', 'Média']].describe()
print(estatisticas)
```

```python
# Número total de alunos
total = len(df)

# Alunos com média > 8
acima_8 = len(df[df['Média'] > 8])

# Probabilidade
probabilidade = acima_8 / total
print(f'Probabilidade de nota média > 8: {probabilidade:.2%}')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Dados
dados = {
    'Aluno': ['Ana', 'Bruno', 'Carlos', 'Diana', 'Eduardo'],
    'Nota_Prova_1': [8.0, 7.5, 9.0, 6.5, 8.5],
    'Nota_Prova_2': [7.0, 8.0, 9.5, 6.0, 9.0]
}

df = pd.DataFrame(dados)
df['Média'] = df[['Nota_Prova_1', 'Nota_Prova_2']].mean(axis=1)

# Estatísticas
media = df['Média'].mean()
desvio = df['Média'].std()
x = np.linspace(media - 3*desvio, media + 3*desvio, 100)
y = stats.norm.pdf(x, media, desvio)

# Gráfico aprimorado
plt.figure(figsize=(10, 6))
plt.hist(df['Média'], bins=5, density=True, alpha=0.6, color='#87CEEB', edgecolor='black', label='Histograma')
plt.plot(x, y, color='darkred', linewidth=2.5, label='Curva Normal')

# Linha da média
plt.axvline(media, color='green', linestyle='--', linewidth=2, label=f'Média = {media:.2f}')

# Anotação da média
plt.text(media + 0.05, max(y)*0.8, f'μ = {media:.2f}\nσ = {desvio:.2f}', color='green', fontsize=10)

# Detalhes do gráfico
plt.title('Distribuição das Notas Médias dos Alunos', fontsize=14, fontweight='bold')
plt.xlabel('Nota Média', fontsize=12)
plt.ylabel('Densidade', fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()



```
